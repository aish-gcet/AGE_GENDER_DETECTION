```python
from datetime import datetime,timedelta

def age_cal(dob_serial,photo_taken):
    if np.isnan(dob_serial) or np.isnan(photo_taken) or dob_serial <= 366:
        return np.nan
    ref_date=datetime(1,1,1)
    dob=ref_date+timedelta(days=int(dob_serial)-366)
    
    photo_taken=datetime(photo_taken,7,1)
    age=photo_taken.year-dob.year

    if dob.month>photo_taken.month or (dob.month==photo_taken.month and dob.day>photo_taken.day):
        age-=1
    return age

    
```

```python
dob_serial=dob_serial.squeeze()
photo_taken=photo_taken.squeeze()
print(dob_serial)
```

```python
ages=[age_cal(d,p) for d,p in zip(dob_serial,photo_taken)]
print(ages)
```

```python
df=pd.DataFrame({
    'ages':ages,
    'photo_year':photo_taken,
    'gender':gender,
    'face_score':face_score,
    'second_face_score':second_face_score,
    'full_path':full_path
    

})
df.head()
```

```python
# 2 df isliye use hue hai jisse poora database aa paye agar do facescore use karte to database se sirf wahi columns select hojate
```

```python
df=df.dropna(subset=['gender','ages','photo_year'])
df=df[df['face_score']>1]
df=df[df['second_face_score'].isna()]
df=df[(df['ages']>=0) & (df['ages']<=100)]
df.head()
```
